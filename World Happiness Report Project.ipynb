```python
import pandas as pd

# Providin url to Load the dataset
url = "https://github.com/FlipRoboTechnologies/ML-Datasets/raw/main/World%20Happiness/happiness_score_dataset.csv"

# Read the data from the URL
data = pd.read_csv(url)

# Display basic information about the dataset
print(data.info())

# Display the first few rows of the dataset
print(data.head())

# Calculate the average happiness score by region
avg_happiness_by_region = data.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)

# Print the average happiness score by region
print(avg_happiness_by_region)

# Calculate the average happiness score by country
avg_happiness_by_country = data.groupby('Country')['Happiness Score'].mean().sort_values(ascending=False)

# Print the average happiness score by country
print(avg_happiness_by_country)

# Calculate the correlation between GDP per capita and happiness score
correlation_gdp = data['Economy (GDP per Capita)'].corr(data['Happiness Score'])

# Print the correlation
print("Correlation between GDP per capita and Happiness Score:", correlation_gdp)

# Calculate the correlation between Healthy life expectancy and happiness score
correlation_health = data['Health (Life Expectancy)'].corr(data['Happiness Score'])

# Print the correlation
print("Correlation between Healthy life expectancy and Happiness Score:", correlation_health)
```
